Premier League Predictor

Najpierw importujemy dane o wynikach w poprzednich sezonach Premier League (od 1993 do obecnego)

```python
import pandas as pd
import numpy as np
import os
```

```python
raw_folder = "../data/raw/pl_seasons/"
processed_folder = "../data/processed/"

csv_files = [f for f in os.listdir(raw_folder) if f.endswith('.csv')]
csv_files.sort()

dataframes_list = []

for file in csv_files:
    file_path = os.path.join(raw_folder, file)
    try:
        df_season = pd.read_csv(file_path, encoding='windows-1252')
        df_season['SeasonFile'] = file
        dataframes_list.append(df_season)
    except Exception as e:
        print("error")
        
df = pd.concat(dataframes_list, ignore_index = True)

```
